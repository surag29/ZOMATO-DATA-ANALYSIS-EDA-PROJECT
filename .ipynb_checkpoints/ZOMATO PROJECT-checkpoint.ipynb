ZOMATO DATA ANALYSIS PROJECT

STEP 1 : IMPORTING LIBRARIES

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

STEP 2 : CREATE THE DATAFRAME 

```python
dataframe = pd.read_csv('Zomato data .csv')
```

```python
dataframe
```

CONVERT THE DATA TYPE OF THE COLUMN - RATE 

```python
def handleRate(value):
    value = str(value).split('/')
    value = value[0]
    return float(value)
dataframe['rate'] = dataframe['rate'].apply(handleRate)

dataframe
```

UNDERSTANDING THE DATA - MISSING DATA

```python
dataframe.info()
```

```python
dataframe.isnull().sum()
```

BUISNESS UNDERSTANDING 

WHAT TYPE OF RESTAURANT DO THE MAJORITY OF THE CUSTOMERS ORDER FROM 

```python
dataframe.head()
```

```python
gx = sns.countplot(x = dataframe['listed_in(type)'])
for bars in gx.containers:
    gx.bar_label(bars)
    
plt.xlabel('type of resturant')
```

CONCLUSION - MAJORITY OF THE RESTURANT FALLS IN DINING CATEGORY

HOW MANY VOTES HAS EACH TYPE OF RESTURANT RECEIVED FROM CUSTOMERS

```python
dataframe.head()
```

```python
groupdata = dataframe.groupby('listed_in(type)')['votes'].sum()
result = pd.DataFrame({'votes': groupdata})
plt.plot(result , c = "green",marker = "o")
plt.xlabel('TYPE OF RESTURANT', c = 'red' , size = 20)
plt.ylabel('votes', c = 'red' , size = 20)
```

CONCLUSION - DINNING RESTURANTS HAS RECEIVED MAXIMUM VOTES

WHAT ARE THE RATINGS THAT MAJORITY OF THE RESTURANT HAS RECIEVED

```python
dataframe.head()
```

```python
fig, ax = plt.subplots()

hx = ax.hist(dataframe['rate'], bins=5)

for bars in ax.containers:
    ax.bar_label(bars)

ax.set_title('Ratings Distribution')
plt.show()
```

CONCLUSIN - THE MAJORITY RESTURANTS  RECEIVED RATINGS FROM 3.5 TO 4

ZOMATO OBSERVED THAT MOST OF THE COUPLES ORDER FOOD ONLINE THEN WHAT IS THEIR AVERAGE SPENDING ON EACH OTHER 

```python
dataframe.head()
```

```python
couple_data = dataframe['approx_cost(for two people)']
sns.countplot(x = couple_data)
```

CONCLUSION -  THE MAJORITY OF COUPLES PREFER RESTUARNT WTH AN APPROXIMATE COST OF 300 RUPEES

WHICH MODE RECEIVES MAXIMUM RATING

```python
dataframe.head()
```

```python
plt.figure(figsize=(6,6))
sns.boxplot(x = 'online_order' ,  y = 'rate', data = dataframe)
```

CONCLUSION -  OFFLINE RATING RECEIVED LOWER RATING IN COMAPRISON TO ONLINE ORDER

```python
dataframe.head()
```

```python
pivot_table = dataframe.pivot_table(
    index='listed_in(type)',
    columns='online_order',
    aggfunc='size',
    fill_value=0
)

sns.heatmap(pivot_table, annot=True, cmap='YlGnBu', fmt='d')

plt.xlabel('online_order')
plt.ylabel('listed_in(type)')
plt.show()
```

CONCLUSION - DINNING RESTURANTS PRIMARILY ACCEPT OFFLINE ORDERS WHEREAS CAFES PRIMARILY RECEIVED ONLINE ORDERS THIS SUGGESTS THAT CLIENT PREFERS ORDERS IN PERSON AT RESTAURANTS BUT PREFER ONLINE ORDERING AT CAFES
